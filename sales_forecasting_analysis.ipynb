```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(
    r"C:\Users\User\Desktop\sales-forecasting-analysis\data\Sample - Superstore.csv",
    encoding="latin1"
)

df.head()
```

```python
df.info()
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])

df["Order Date"].head()
```

```python
monthly_sales=df.groupby(pd.Grouper(key="Order Date", freq="ME"))["Sales"].sum()
monthly_sales.head()
```

```python
ax=monthly_sales.plot(figsize=(12,6))
ax.set_title("Monthly Trend Sales Analysis")
ax.set_xlabel("Oder Date")
ax.set_ylabel("Total Sales")
```

```python
forecast_trend = monthly_sales.rolling(window=3).mean()
ax = monthly_sales.plot(figsize=(12,6),label="Monthly_sales")

forecast_trend.plot(ax=ax, label="Forecast_Trend")

ax.set_title("Sales Forecasting Trend Analysis")

ax.set_xlabel("Order Date")
ax.set_ylabel("Total Orders")

plt.legend()
```
